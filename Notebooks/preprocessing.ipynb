```python
import pandas as pd

# Load Data-set
url = 'https://raw.githubusercontent.com/ThaLovelace/CS372_Mini-Project-1/refs/heads/main/Data/Sleep_health_and_lifestyle_dataset.csv'
df = pd.read_csv(url)
```

```python
# Check missing values before imputation
print("Missing values before:")
display(df.isnull().sum())
```

```python
# Impute missing values in 'Sleep Disorder' column with 'None'
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None')

# Check missing values after imputation
print("\nMissing values after:")
display(df.isnull().sum())
```

```python
# Display columns before dropping 'Person ID'
print("Columns before dropping:")
display(df.columns)
```

```python
# Drop 'Person ID' column
df = df.drop('Person ID', axis=1)

# Display columns after dropping
print("Columns after dropping:")
display(df.columns)
```

```python
# Display 'Blood Pressure' column before splitting
print("Blood Pressure before split:")
display(df[['Blood Pressure']].head())
```

```python
# Split into 'Systolic_BP' and 'Diastolic_BP', then convert to integers
df[['Systolic_BP', 'Diastolic_BP']] = df['Blood Pressure'].str.split('/', expand=True)
df['Systolic_BP'] = df['Systolic_BP'].astype(int)
df['Diastolic_BP'] = df['Diastolic_BP'].astype(int)

# Drop the old 'Blood Pressure' column
df = df.drop('Blood Pressure', axis=1)

# Display new columns
print("Blood Pressure after split:")
display(df[['Systolic_BP', 'Diastolic_BP']].head())
```

```python
# Display unique categories before consistency fix
print("BMI Categories before fix:")
display(df['BMI Category'].value_counts())
```

```python
# Replace 'Normal Weight' with 'Normal'
df['BMI Category'] = df['BMI Category'].replace('Normal Weight', 'Normal')

# Display unique categories after consistency fix
print("BMI Categories after fix:")
display(df['BMI Category'].value_counts())
```

```python
# Display BMI before encoding
print("BMI before Ordinal Encoding:")
display(df[['BMI Category']].head())
```

```python
# Apply Ordinal Encoding (Normal=0, Overweight=1, Obese=2)
bmi_mapping = {'Normal': 0, 'Overweight': 1, 'Obese': 2}
df['BMI Category'] = df['BMI Category'].map(bmi_mapping)

# Display BMI after encoding
print("BMI after Ordinal Encoding:")
display(df[['BMI Category']].head())
```

```python
# Display Gender and Occupation before encoding
print("Features before One-Hot Encoding:")
display(df[['Gender', 'Occupation']].head())
```

```python
# Apply One-Hot Encoding for 'Gender' and 'Occupation'
df = pd.get_dummies(df, columns=['Gender', 'Occupation'])

# Display the dataset after One-Hot Encoding
print("Features after One-Hot Encoding:")
display(df.head())
```

```python
# นำเข้าเครื่องมือสำหรับทำ Standardization
from sklearn.preprocessing import StandardScaler

# กำหนดคอลัมน์ที่เป็นตัวเลข
num_cols = ['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 
            'Stress Level', 'Heart Rate', 'Daily Steps', 'Systolic_BP', 'Diastolic_BP']

# เช็คค่าสถิติก่อนทำ
print("Data before scaling:")
display(df[num_cols].describe().loc[['mean', 'std', 'min', 'max']])
```

```python
# Apply Standardization (Z-score scaling)
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

# Display statistical summary after scaling
print("Data after scaling (Mean is ~0, Std is 1):")
display(df[num_cols].describe().loc[['mean', 'std', 'min', 'max']].round(2))
```

```python
# Save the cleaned and preprocessed dataset to a new CSV file
df.to_csv('cleaned_sleep_data.csv', index=False)
print("Data preprocessing complete. Saved as 'cleaned_sleep_data.csv'")
```
